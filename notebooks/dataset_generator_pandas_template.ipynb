```python
import pandas as pd
import numpy as np
```

#### Create our dataset

```python
def generate_data(size = 10_000):
  df = pd.DataFrame()
  df['age'] = np.random.randint(18, 65, size)
  df['time_in_bed'] = np.random.randint(0, 9, size)
  df['favourite_food'] = np.random.choice(['pizza', 'ice cream', 'fish and sea food', 'chicken', 'pasta', 'rice', 'vegetables', 'meat'], size)
  return df
```

```python
df = generate_data()
```
