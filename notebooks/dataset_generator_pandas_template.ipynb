**Import Libraries**

```python
import pandas as pd
import numpy as np
import polars as pl
```

**Generate our dataset**

```python
def generate_data(size = 10_000):
  df = pd.DataFrame()
  df['age'] = np.random.randint(18, 65, size)
  df['time_in_bed'] = np.random.randint(0, 9, size)
  df['favourite_food'] = np.random.choice(['pizza', 'ice cream', 'fish and sea food', 'chicken', 'pasta', 'rice', 'vegetables', 'meat'], size)
  return df

df = generate_data()
```

**Convert to Polars Dataframe & perform group by Aggregation**

```python
df_polars = pl.from_pandas(df)

df_polars_agg = df_polars.group_by('favourite_food').agg(
    pl.col('favourite_food').count().alias('total_count')
)

df_polars_agg
```
